```python
import os
import pprint
import json
import numpy as np
```

```python
files = os.listdir( ) ## save the names of each json file in the folder
```

```python
for js in files: ## for each json annotation file
    
    data = json.load( open( js ) ) ## load the file
    
    ## check if any bounding boxes have both a None in the frame_count column and a none in the frame_number column
    if sum( np.array( list( map( lambda x: x[ 'frame_count' ] is None and x[ 'frame_number' ] is None, data ) ) ) ) != 0:
        
        ## flag any problematic annotation files
        print( js + ' ERROR: EITHER FRAME COUNT OR FRAME NUMBER IS NONE' )
        
    else:
        
        print( js + ' is okay' )

```
